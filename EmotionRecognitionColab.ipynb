```
import tensorflow as tf
import sklearn
import sklearn.linear_model
import sklearn.datasets
import matplotlib.pyplot as plt
import sklearn.ensemble
import pickle
import pylab as pl
import scipy.misc as smp
import pandas as pd
import numpy as np
import os
from PIL import Image
import math
import imghdr
import cv2
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Activation,Flatten,Dropout
from functools import partial
from operator import is_not
from keras.preprocessing.image import img_to_array,load_img
from PIL import Image

```

```
def labelsandimages(root_dir):
    file_dir = {}
    parent_folder = [os.path.join(root_dir,f) for f in os.listdir(root_dir)]
    for subfolder in parent_folder:
      folders = [os.path.join(subfolder,f) for f in os.listdir(subfolder)]
        
      for subfolder1 in folders:
        if(not os.path.isdir(subfolder1)):
          continue
        filecontainer = [os.path.join(subfolder1,f) for f in os.listdir(subfolder1)]
        filecontainer.sort()
        
        for file in filecontainer:
              parent_key = file.split('/')[-3] + file.split('/')[-2]
              if(parent_key not in file_dir):
                file_dir[parent_key]={}
              if(file.endswith(".txt")):
                file_dir[parent_key]['text_file'] = file
              elif(file.endswith(".png")):
                if('image' not in file_dir[parent_key]):
                  file_dir[parent_key]['image'] = []              
                file_dir[parent_key]['image'].append(file)
            
    return file_dir
            
```

```
image_and_label = labelsandimages("/content/drive/My Drive/Dataset/Emotion/cohn-kanade-images")
```

```
image=np.zeros((1635,200,200,3),dtype = object)
label=np.zeros((1635,),dtype=int)

      

    


    
```

```
count_y=0
count_x=0
for d_id,value in image_and_label.items():
  for key in value:
    if('text_file' in key):
      f = open(value['text_file'],"rb")
      y = int(float(f.readline()))
      for i in range(4):
        label[count_y] =y
        count_y+=1
      img = cv2.imread(value['image'][-1])
      img = cv2.resize(img,(200,200),3)
      image[count_x]=img
      count_x+=1
      img = cv2.imread(value['image'][-2])
      img = cv2.resize(img,(200,200),3)
      image[count_x]=img
      count_x+=1
      img = cv2.imread(value['image'][-3])
      img = cv2.resize(img,(200,200),3)
      image[count_x]=img
      count_x+=1
      img = cv2.imread(value['image'][-4])
      img = cv2.resize(img,(200,200),3)
      image[count_x]=img
      count_x+=1
      label[count_y] = 0
      count_y+=1
      img = cv2.imread(value['image'][0])
      img = cv2.resize(img,(200,200),3)
      image[count_x]=img
      count_x+=1
```

```
labels = label.reshape(-1,1)
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(categorical_features =[0])
labels = onehotencoder.fit_transform(labels).toarray()
```

```
labels.shape
for i in range(10):
    print(labels[i])
```

```
image , labels = shuffle(image,labels)
x_train,x_test,y_train,y_test = train_test_split(image,labels,test_size=0.2) 
```

```
from keras.applications import resnet50
```

```
base_model = resnet50.ResNet50(
    weights = 'imagenet', include_top=False, input_shape=(200,200, 3))
```

```
for layer in base_model.layers[:-4]:
    layer.trainable = False
```

```
for layer in base_model.layers:
    print(layer, layer.trainable)
base_model.summary()
```

```
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.6))
model.add(Dense(200, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(8, activation='softmax'))
```

```
model.summary()
```

```
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['acc'])
history = model.fit(x_train,y_train,epochs=100,
                   validation_data=(x_test,y_test),
                   verbose = 1)
```

```
"""Let's test our model with more generated data"""
cross_loss, cross_acc = model.evaluate(x_test,y_test)
train_loss, train_acc = model.evaluate(x_train,y_train)
predictions = np.argmax(model.predict(x_train),axis=1)
print("Train Accuracy",train_acc)
print("Cross Validation Accuracy",cross_acc)
```
